```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
import pandas as pd
from sklearn.metrics import r2_score
import math


# #the files containing data scraped from graphs
filename50 = 'stakebake_data/stakebake50S.csv'
# #read in data for each temperature
dat50 = pd.read_csv(filename50)

#mass U not given but needed for equation
#can be calculated from the total hydrogen reacted maxh.
def calc_U_mass(maxh):
    atoms = maxh*0.5*(6.022e23)
    atoms_in_cm3 = 8.22e22
    U_cm3 = atoms/atoms_in_cm3
    rho_U = 19
    mass_U = rho_U*U_cm3
    return mass_U

#performing the Stakbake equation
arrk = 4.84
wm = 11.82
T = 323
n=0.65
K = arrk*(n-1)
Mo = calc_U_mass(wm)
M = Mo**(n-1)
#array of times between 0 and M_0^(1-n)/K(1-n), resolution=0.1
upper_t_lim = (Mo**(1-n))/(arrk*(1-n))
upper_t_lim = math.floor(upper_t_lim) #for all t>this wt = wm so can extend plot if want
print(upper_t_lim)
t=np.linspace(0,upper_t_lim,(upper_t_lim) * 10 +1)


wt = (1-(1+(K*M*t))**(1/(1-n)))*wm

plt.plot(t, wt)
```
