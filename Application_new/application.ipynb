```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import gridplot
from bokeh.io import output_notebook, push_notebook
from time import sleep
from bokeh.models import Arrow, NormalHead, Label


TIME=TIME
# Set up the notebook for inline plotting
output_notebook()

# Create some sample data
x = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Create Bokeh figures with fixed axis ranges
p1 = figure(title="Plot 1", plot_width=300, plot_height=300, x_range = (0,10e6), y_range=(0,1.3))

p2 = figure(width=300, height=300, title="Plot 2", toolbar_location=None, x_range=(-100000, 8000000), y_range=(-5, -1))
p3 = figure(width=300, height=300, title="Plot 3", toolbar_location=None,x_range = (-1,1))

centre_x, centre_y=0,0

# Create ColumnDataSources for updating data
source1 = ColumnDataSource(data=dict(x=[], y=[]))
source1b = ColumnDataSource(data=dict(x=[], y=[]))

source2 = ColumnDataSource(data=dict(x=[], y=[]))
source3 = ColumnDataSource(data=dict(x=[], y=[]))
source4 = ColumnDataSource(data=dict(x=[0], y=[0], radius=[1]))
source5 = ColumnDataSource(data=dict(x=[0], y=[0], radius=[1]))
source6 = ColumnDataSource(data=dict(x=[1], y=[0]))




# Create line glyphs
line1 = p1.line('x', 'y', source=source1, line_width=2, line_color="blue")
line1b = p1.line('x', 'y', source=source1b, line_width=2, line_color="red")

line2 = p2.line('x', 'y', source=source2, line_width=2, line_color="orange")
#line3 = p3.line('x', 'y', source=source3, line_width=2, line_color="green")
circle = p3.circle('x', 'y', size='radius', source=source4, line_color="blue", fill_color="blue")
circle2 = p3.circle('x', 'y', size='radius', source=source5, line_color="red", fill_color="red")
line = p3.line(x='x', y='y', line_width=2, line_color="orange", source=source6)
                           


# Create a layout with the three plots in a horizontal grid
layout = gridplot([[p1, p2, p3]])

# Display the plot in the notebook
handle = show(layout, notebook_handle=True)
# Update function for animation
def update():
    for frame in range(0, len(TIME)):
        source1.data = dict(x=TIME[:frame], y=ORAD[:frame])
        source1b.data = dict(x=TIME[:frame], y=CRAD[:frame])

        source2.data = dict(x=TIME[:frame], y=PRES[:frame])
        source3.data = dict(x=TIME[:frame], y=ORAD[:frame])
        source4.data = dict(x=[0], y=[0], radius=[ORAD[frame]*100])
        source5.data = dict(x=[0], y=[0], radius=[CRAD[frame]*100])
        source6.data = dict(x=[0,CRAD[frame]/2], y=[0, 0])
                           


        # Push the updated plot to the notebook
        push_notebook(handle=handle)

        # Pause for a short duration to slow down the animation
        sleep(0.1)

# Manually create the animation
update()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from IPython.display import display, clear_output

# Create a figure
p = figure(width=400, height=400, title="Animated Horizontal Line")

# Set the range for the y-axis
p.y_range.start = 0
p.y_range.end = 10

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=[2], y=[5]))

# Plot a horizontal line using the data from the ColumnDataSource
line = p.line(x='x', y='y', line_width=2, line_color="red", source=source)

# Show the plot with a specific display_id for Jupyter Notebook inline plotting
display_id = show(p, notebook_handle=True)

# Update function for animation
def update_b(new_b):
    source.data = dict(x=[2, new_b], y=[5, 5])

# Manually create the animation by updating b
for new_b in range(3, 90):
    update_b(new_b)
    
    # Clear the previous output and display the updated plot
    clear_output(wait=True)
    show(p, notebook_handle=True, display_id=display_id)


```
